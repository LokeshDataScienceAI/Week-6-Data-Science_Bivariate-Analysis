```python
import matplotlib.pyplot as plt

# Select numerical columns
dataset_num = dataset.select_dtypes(include=["int64","float64"])

# Pairplot with regression line
g = sns.pairplot(dataset_num, kind="reg", diag_kind="kde")

# Add Positive / Negative / Zero text
for i, row in enumerate(dataset_num.columns):
    for j, col in enumerate(dataset_num.columns):

        if i != j:
            corr = dataset_num[row].corr(dataset_num[col])

            if corr > 0:
                text = "Positive"
            elif corr < 0:
                text = "Negative"
            else:
                text = "Zero"

            g.axes[i,j].text(0.5, -0.3,
                             text,
                             fontsize=10,
                             color="blue",
                             ha="center",
                             transform=g.axes[i,j].transAxes)

plt.show()
```

Calculate Variance Inflation Factor - Fix Multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i)for i in range(X.shape[1])]
    
    return vif
```

```python
calc_vif(dataset_num)
```
