```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot with Regression Line

sns.regplot(x="etest_p",
            y="mba_p",
            data=dataset,
            scatter_kws={"color":"red"},
            line_kws={"color":"blue"})

plt.xlabel("etest_p", color="purple", fontsize=12)
plt.ylabel("mba_p", color="green", fontsize=12)

plt.title("Covariance")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
dataset_num = dataset.select_dtypes(include=["int64","float64"])

# Covariance Matrix
cov_matrix = dataset_num.cov()

# Covariance Heatmap
plt.figure(figsize=(10,5))
sns.heatmap(cov_matrix, annot=True, cmap="YlGnBu")

plt.title("Covariance Heatmap")
plt.show()
```
