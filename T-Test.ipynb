Calculate Variance Inflation Factor - Fix Multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i)for i in range(X.shape[1])]
    
    return vif
```

```python
calc_vif(dataset_num)
```

Independent Sample - Unpaired T Test

Different group (Central and Others) but same condition(hsc_p)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Central=dataset[dataset['hsc_b']=='Central']['salary']
Others=dataset[dataset['hsc_b']=='Others']['salary']
ttest_ind(Central, Others)
```

Dependent Sample-Paired T_Test

Same Group(Commerce) but Different Condition(etest_p, mba_P)

```python
from scipy.stats import ttest_rel

# Remove null values
dataset = dataset.dropna()

# Filter Commerce students
commerce = dataset[dataset['hsc_s'] == 'Commerce']

# Two conditions
etest = commerce['etest_p']
mba = commerce['mba_p']

# Paired T-Test
ttest_rel(etest, mba)
```
