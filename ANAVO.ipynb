```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i)for i in range(X.shape[1])]
    
    return vif
```

```python
#Independent Sample - Unpaired T Test
#Different group (Central and Others) but same condition(hsc_p)
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Central=dataset[dataset['hsc_b']=='Central']['salary']
Others=dataset[dataset['hsc_b']=='Others']['salary']
ttest_ind(Central, Others)
```

```python
#Dependent Sample-Paired T_Test
#Same Group(Commerce) but Different Condition(etest_p, mba_P)
from scipy.stats import ttest_rel

# Remove null values
dataset = dataset.dropna()

# Filter Commerce students
commerce = dataset[dataset['hsc_s'] == 'Commerce']

# Two conditions
etest = commerce['etest_p']
mba = commerce['mba_p']

# Paired T-Test
ttest_rel(etest, mba)
```

```python
from scipy.stats import ttest_rel

# Selecting only Commerce students
commerce = dataset[dataset['hsc_s'] == 'Commerce']

# Same students two conditions
etest = commerce['etest_p']
mba = commerce['mba_p']

# Paired T-Test
stat, p_value = ttest_rel(etest, mba)

# Output
print("T-Test Value:", stat)
print("P-Value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("There is a significant difference between etest_p and mba_p scores")
else:
    print("Accept Null Hypothesis")
    print("There is NO significant difference between etest_p and mba_p scores")
```

Two way ANAVO

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Remove salary = 0
dataset = dataset[dataset['salary'] > 0]

# Two-Way ANOVA
model = ols('salary ~ C(gender) * C(specialisation)', 
            data=dataset).fit()

result = sm.stats.anova_lm(model, typ=2)

print(result)
```

A Two-Way ANOVA was conducted to examine the effect of gender and
specialisation on salary.

The analysis showed that gender did not have a statistically significant
effect on salary, F = 3.483, p = 0.064.

Specialisation also did not show a statistically significant effect on
salary, F = 2.785, p = 0.097.

Additionally, there was no significant interaction effect between gender
and specialisation on salary, F = 0.066, p = 0.798.

Since all p-values were greater than the significance level (0.05),
the null hypotheses were accepted.
